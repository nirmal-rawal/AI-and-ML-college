```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1, n_targets=1, noise=80,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
#let apply OLS
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green')
```

# Lets apply  Gradient Descent assuming slope is constant m=78.35
# and lets assum the starting values of intercept b=0

```python
y_pred=((78.35*X+0).reshape(4))
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green',label="OLS")
plt.plot(X,y_pred,color='red',label="b=0")
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loos_slop=-2*np.sum(y-m*X.ravel()-b)
loos_slop
```

```python
#lets take a learning rate
l=0.1
step_size=l*loos_slop
step_size
```

```python
#calculate the new intercept
b=b- step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green',label="OLS")
plt.plot(X,y_pred,color='red',label="b=0")
plt.plot(X,y_pred1,color='black',label="b=20.92")
plt.legend()
plt.show()
```

```python
step_size=l*loos_slop
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green',label="OLS")
plt.plot(X,y_pred,color='red',label="b=0")
plt.plot(X,y_pred1,color='black',label="b=20.92")
plt.plot(X,y_pred2,color='pink',label="b=25.113")
plt.legend()
plt.show()
```

# Gradient Descents using **loop**

```python
b=-100
m=78.35
lr=0.1#learning rate

epochs=10

for i in range(epochs):
  loos_slop=-2*np.sum(y-m*X.ravel()-b)
  b=b-(lr*loos_slop)
  y_predict=m*X+b
  plt.plot(X,y_predict)


plt.scatter(X,y)

```
